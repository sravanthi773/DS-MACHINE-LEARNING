```python
import numpy as np
from sklearn.model_selection import train_test_split
x=np.arange(20).reshape(2,10).T
y=np.arange(10)
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)

print(x_train)
print()
print(x_test)
print()
print(y_train)
print()
print(y_test)
```
